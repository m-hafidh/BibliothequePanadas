```python
#Cellules vides
#Les cellules vides peuvent potentiellement vous donner un résultat erroné lorsque vous analysez des données.
#Supprimer des lignes

df = pd.read_csv('data.csv')

new_df = df.dropna()

print(df.to_string())
```

```python
#Si vous souhaitez modifier le DataFrame d'origine, utilisez l' inplace = Trueargument :
df = pd.read_csv('data.csv')
df.dropna(inplace= True)

print(df.to_string())
```

```python
#La fillna()méthode nous permet de remplacer les cellules vides par une valeur :
df=pd.read_csv('data.csv')
df.fillna(130, inplace =True)
print(df.to_string())
```

```python
#Remplacez les valeurs NULL dans les colonnes "Calories" par le nombre 130 :Remplacez les valeurs NULL dans les colonnes "Calories" par le nombre 160 :
df=pd.read_csv('data.csv')
df.fillna(160, inplace =True)
print(df.to_string())
```

```python
#Calculez la MOYENNE et remplacez toutes les valeurs vides par celle-ci :
df = pd.read_csv('data.csv')

x = df["Calories"].mean()

df["Calories"].fillna(x, inplace = True)
print (df.to_string())
```

```python
#Calculez la MÉDIANE et remplacez toutes les valeurs vides par celle-ci :
df = pd.read_csv('data.csv')

x = df["Calories"].median()

df["Calories"].fillna(x, inplace = True)
print (df.to_string())
```

```python
#Calculez le MODE et remplacez toutes les valeurs vides par celui-ci :
df = pd.read_csv('data.csv')

x = df["Calories"].mode()[0]

df["Calories"].fillna(x, inplace = True)
print (df.to_string())
```

```python
#Nettoyage des données de mauvais format
#Convertir en date :en utilisant la methode "to_datetime()"
df = pd.read_csv('data.csv')

df['Date'] = pd.to_datetime(df['Date'])

print(df.to_string())
```

```python
#Supprimez les lignes avec une valeur NULL dans la colonne "Date" :
df.dropna(subset=['Date'], inplace = True)
```

```python
#Données erronées
#Remplacement des valeurs
df.loc[7, 'Duration'] = 45
print (df.to_string())


```

```python
#Parcourez toutes les valeurs de la colonne "Durée".

#Si la valeur est supérieure à 120, définissez-la sur 120 :
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
print (df.to_string())
```

```python
#Supprimez les lignes dont la "Durée" est supérieure à 120 :
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
print(df.to_string())
```

```python
#Suppression des doublons
#La duplicated()méthode renvoie des valeurs booléennes pour chaque ligne :
#Renvoie Truepour chaque ligne qui est un doublon, sinonFalse :
print(df.duplicated())


```

```python
#Pour supprimer les doublons, utilisez la drop_duplicates()méthode.
df.drop_duplicates(inplace = True)
print (df.to_string())
```

```python
#Corrélations de données
df.corr()
```

Nuage de points

Spécifiez que vous souhaitez un nuage de points avec l' kindargument :

kind = 'scatter'

Un nuage de points a besoin d’un axe x et d’un axe y.

Dans l'exemple ci-dessous, nous utiliserons « Durée » pour l'axe des x et « Calories » pour l'axe des y.

Incluez les arguments x et y comme ceci :

x = 'Duration', y = 'Calories'


```python
df.plot(kind = 'scatter', x = 'Duration', y = 'Calories')

plt.show()
```

```python
#Un nuage de points où il n'y a aucune relation entre les colonnes :

df.plot(kind = 'scatter', x = 'Duration', y = 'Maxpulse')

plt.show()
```

                                Histogramme

Utilisez l' kindargument pour spécifier que vous souhaitez un histogramme :

kind = 'hist'

Un histogramme n'a besoin que d'une seule colonne.

Un histogramme nous montre la fréquence de chaque intervalle, par exemple combien d'entraînements ont duré entre 50 et 60 minutes ?

Dans l'exemple ci-dessous nous utiliserons la colonne "Durée" pour créer l'histogramme :

```python
df["Duration"].plot(kind = 'hist')
```
